### Drawing Circuit Elements
To start using Schemdraw, import the library:
```python

from schemdraw import elements as e
```

Schemdraw provides a variety of components to create circuit diagrams. Below are the most common elements:

- **Resistor:** `e.Resistor()`
- **Capacitor:** `e.Capacitor()`
- **Inductor:** `e.Inductor()`
- **Voltage Source:** `e.SourceV()`
- **Ground:** `e.Ground()`

#### Example:
Elements inside a `with` block are automatically added to the drawing. The drawing maintains a current position and direction, such that the next element will start at the end of the previous element, following the same direction.

```python
import schemdraw
from schemdraw import elements as e

with schemdraw.Drawing() as d:
    d += e.Capacitor()
    d += e.Resistor()
    d += e.Diode()
```

#### Changing Direction:
You can rotate elements using direction methods like `.up()`, `.down()`, `.left()`, and `.right()`. The subsequent elements will follow the same direction.

```python
import schemdraw
from schemdraw import elements as e

with schemdraw.Drawing() as d:
    d += e.Capacitor()
    d += e.Resistor().up()  # Rotates the Resistor upwards
    d += e.Diode()
```

Example 1: Simple Series Circuit
Target: Create a circuit with a voltage source, two resistors in series, and ground.

````python
import schemdraw
from schemdraw import elements as e

with schemdraw.Drawing() as d:
    d += e.SourceV().label('V1')  # Voltage source with label V1
    d += e.Resistor().right().label('R1')  # First resistor with label R1
    d += e.Resistor().right().label('R2')  # Second resistor with label R2
    d += e.Ground()  # Ground connection
````

Example 2: RC Circuit
Target: Create a simple RC circuit with a voltage source, one resistor, one capacitor, and ground.



````python
import schemdraw
from schemdraw import elements as e

with schemdraw.Drawing() as d:
    d += e.SourceV().label('V1')  # Voltage source labeled V1
    d += e.Resistor().right().label('R1')  # Resistor labeled R1
    d += e.Capacitor().down().label('C1')  # Capacitor labeled C1, placed vertically
    d += e.Ground()  # Ground connection
````

Example 3: Parallel Circuit
Target: Create a parallel circuit with two resistors connected in parallel to a voltage source and ground.



````python
import schemdraw
from schemdraw import elements as e

with schemdraw.Drawing() as d:
    d += e.SourceV().label('V1')  # Voltage source
    d += (e.Line().right() + e.Dot().at([0,0]))  # Connecting line with a node
    
    # Resistor R1 path
    d += e.Resistor().down().at([1,0]).label('R1')  # Resistor R1 vertically down
    d += e.Ground().at([1, -1])  # Ground connected to Resistor R1
    
    # Resistor R2 path
    d += e.Resistor().right().at([0,0]).label('R2')  # Resistor R2 horizontally right
    d += e.Ground().at([1, 0])  # Ground connected to Resistor R2
````

Example 4: Circuit with Direction Changes
Target: Create a circuit where components change directions (up/down) and are labeled.



````python
import schemdraw
from schemdraw import elements as e

with schemdraw.Drawing() as d:
    d += e.SourceV().label('V1')  # Voltage source labeled V1
    d += e.Resistor().right().label('R1')  # Resistor R1 to the right
    d += e.Capacitor().down().label('C1')  # Capacitor C1 vertically down
    d += e.Resistor().left().label('R2')  # Resistor R2 to the left
    d += e.Ground()  # Ground connection
````

Example 5: Annotating the Circuit
Target: Annotate parts of a circuit, like adding a text box for a voltage source and resistor.



````python
import schemdraw
from schemdraw import elements as e

with schemdraw.Drawing() as d:
    d += e.SourceV().label('V1')  # Voltage source labeled V1
    d += e.Resistor().right().label('R1')  # Resistor labeled R1
    d += e.Ground()  # Ground connection
    
    # Adding text annotations
    d += e.Text('Power Supply', loc='left').at([-1, 0.5])  # Annotation for voltage source
    d += e.Text('Resistor Load', loc='right').at([2, 0])  # Annotation for resistor
````
